```python
import requests
import os

# ===== CONFIG =====
ACCESS_KEY = "97zjnfg3Gln9sJSQNhh0GSzZodtRbATFTPjAbEE7Zis"
query = "russian girl"   # change to "books", "coffee", etc.
total_images = 100
per_page = 30   # max allowed by Unsplash
folder = query

# Create folder
os.makedirs(folder, exist_ok=True)

downloaded = 0
page = 1

while downloaded < total_images:
    url = f"https://api.unsplash.com/search/photos"
    params = {
        "query": query,
        "client_id": ACCESS_KEY,
        "per_page": per_page,
        "page": page
    }

    response = requests.get(url, params=params)
    data = response.json()

    for photo in data['results']:
        if downloaded >= total_images:
            break
        
        img_url = photo['urls']['regular']
        img_data = requests.get(img_url).content
        
        with open(f"{folder}/{query}_{downloaded+1}.jpg", "wb") as f:
            f.write(img_data)
        
        downloaded += 1
        print(f"Downloaded {downloaded}")

    page += 1

print("Done!")
```
